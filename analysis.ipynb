```python
import numpy as np

# Define a 2x3 Matrix (e.g., 2 houses, 3 features each)
matrix_A = np.array([
    [2, 4, 6],
    [1, 3, 5]
])

# Define a 3x1 Matrix (e.g., 3 model weights for those features)
matrix_B = np.array([
    [0.5],
    [1.0],
    [2.0]
])

# Perform Matrix Multiplication using NumPy's built-in @ operator
# (2x3) multiplied by (3x1) results in a (2x1) matrix of predictions
result = matrix_A @ matrix_B

print("Prediction Results:\n", result)
```

```python
import numpy as np

# 1. Define the function and its derivative (its gradient)
def cost_function(x):
    return x**2

def get_gradient(x):
    return 2 * x  # The derivative of x^2

# 2. Pick a random starting point and set our step size
current_x = 10.0      # Our "blindfolded guess" starting point
learning_rate = 0.1   # The step multiplier (how big our steps are)
iterations = 40       # How many times we repeat the loop down the mountain

print(f"Starting at x = {current_x}, where cost = {cost_function(current_x)}\n")

# 3. The Gradient Descent Loop
for step in range(iterations):
    # Feel the slope under our feet
    slope = get_gradient(current_x)
    
    # Take a step down the mountain: New = Old - (Learning Rate * Slope)
    current_x = current_x - (learning_rate * slope)
    
    # Print the progress every 5 steps so we can watch it drop
    if step % 5 == 0 or step == iterations - 1:
        print(f"Step {step+1}: x = {current_x:.4f}, cost = {cost_function(current_x):.4f}")

print(f"\nFinal Result: The model found the minimum at x = {current_x:.6f}!")
```
